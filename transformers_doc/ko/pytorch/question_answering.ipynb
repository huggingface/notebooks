## ì „ì²˜ë¦¬[[preprocess]]

````
#@title
from IPython.display import HTML

HTML('<iframe width="560" height="315" src="https://www.youtube.com/embed/qgaM0weJHpA?rel=0&amp;controls=0&amp;showinfo=0" frameborder="0" allowfullscreen></iframe>')
````

ë‹¤ìŒ ë‹¨ê³„ì—ì„œëŠ” `question` ë° `context` í•­ëª©ì„ ì²˜ë¦¬í•˜ê¸° ìœ„í•´ DistilBERT í† í¬ë‚˜ì´ì €ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤:

````
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
````

ì§ˆì˜ ì‘ë‹µ íƒœìŠ¤í¬ì™€ ê´€ë ¨í•´ì„œ íŠ¹íˆ ìœ ì˜í•´ì•¼í•  ëª‡ ê°€ì§€ ì „ì²˜ë¦¬ ë‹¨ê³„ê°€ ìˆìŠµë‹ˆë‹¤:

1. ë°ì´í„° ì„¸íŠ¸ì˜ ì¼ë¶€ ì˜ˆì œì—ëŠ” ëª¨ë¸ì˜ ìµœëŒ€ ì…ë ¥ ê¸¸ì´ë¥¼ ì´ˆê³¼í•˜ëŠ” ë§¤ìš° ê¸´ `context`ê°€ ìˆì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ê¸´ ì‹œí€€ìŠ¤ë¥¼ ë‹¤ë£¨ê¸° ìœ„í•´ì„œëŠ”, `truncation="only_second"`ë¡œ ì„¤ì •í•´ `context`ë§Œ ì˜ë¼ë‚´ë©´ ë©ë‹ˆë‹¤.
2. ê·¸ ë‹¤ìŒ, `return_offset_mapping=True`ë¡œ ì„¤ì •í•´ ë‹µë³€ì˜ ì‹œì‘ê³¼ ì¢…ë£Œ ìœ„ì¹˜ë¥¼ ì›ë˜ì˜ `context`ì— ë§¤í•‘í•©ë‹ˆë‹¤.
3. ë§¤í•‘ì„ ì™„ë£Œí•˜ë©´, ì´ì œ ë‹µë³€ì—ì„œ ì‹œì‘ í† í°ê³¼ ì¢…ë£Œ í† í°ì„ ì°¾ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì˜¤í”„ì…‹ì˜ ì–´ëŠ ë¶€ë¶„ì´ `question`ê³¼ `context`ì— í•´ë‹¹í•˜ëŠ”ì§€ ì°¾ì„ ìˆ˜ ìˆë„ë¡ `sequence_ids` ë©”ì†Œë“œë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.

ë‹¤ìŒì€ `answer`ì˜ ì‹œì‘ í† í°ê³¼ ì¢…ë£Œ í† í°ì„ ì˜ë¼ë‚´ì„œ `context`ì— ë§¤í•‘í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“œëŠ” ë°©ë²•ì…ë‹ˆë‹¤:

````
def preprocess_function(examples):
    questions = [q.strip() for q in examples["question"]]
    inputs = tokenizer(
        questions,
        examples["context"],
        max_length=384,
        truncation="only_second",
        return_offsets_mapping=True,
        padding="max_length",
    )

    offset_mapping = inputs.pop("offset_mapping")
    answers = examples["answers"]
    start_positions = []
    end_positions = []

    for i, offset in enumerate(offset_mapping):
        answer = answers[i]
        start_char = answer["answer_start"][0]
        end_char = answer["answer_start"][0] + len(answer["text"][0])
        sequence_ids = inputs.sequence_ids(i)

        # Find the start and end of the context
        idx = 0
        while sequence_ids[idx] != 1:
            idx += 1
        context_start = idx
        while sequence_ids[idx] == 1:
            idx += 1
        context_end = idx - 1

        # If the answer is not fully inside the context, label it (0, 0)
        if offset[context_start][0] > end_char or offset[context_end][1] < start_char:
            start_positions.append(0)
            end_positions.append(0)
        else:
            # Otherwise it's the start and end token positions
            idx = context_start
            while idx <= context_end and offset[idx][0] <= start_char:
                idx += 1
            start_positions.append(idx - 1)

            idx = context_end
            while idx >= context_start and offset[idx][1] >= end_char:
                idx -= 1
            end_positions.append(idx + 1)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs
````

ëª¨ë“  ë°ì´í„° ì„¸íŠ¸ì— ì „ì²˜ë¦¬ë¥¼ ì ìš©í•˜ë ¤ë©´, ğŸ¤— Datasets `map` í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”. `batched=True`ë¡œ ì„¤ì •í•´ ë°ì´í„° ì„¸íŠ¸ì˜ ì—¬ëŸ¬ ìš”ì†Œë“¤ì„ í•œ ë²ˆì— ì²˜ë¦¬í•˜ë©´ `map` í•¨ìˆ˜ì˜ ì†ë„ë¥¼ ë¹ ë¥´ê²Œ í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. í•„ìš”í•˜ì§€ ì•Šì€ ì—´ì€ ëª¨ë‘ ì œê±°í•©ë‹ˆë‹¤:

````
tokenized_squad = squad.map(preprocess_function, batched=True, remove_columns=squad["train"].column_names)
````

ì´ì œ `DefaultDataCollator`ë¥¼ ì´ìš©í•´ ì˜ˆì‹œ ë°°ì¹˜ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. ğŸ¤— Transformersì˜ ë‹¤ë¥¸ ë°ì´í„° ì½œë ˆì´í„°(data collator)ì™€ ë‹¬ë¦¬, `DefaultDataCollator`ëŠ” íŒ¨ë”©ê³¼ ê°™ì€ ì¶”ê°€ ì „ì²˜ë¦¬ë¥¼ ì ìš©í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤:

````
from transformers import DefaultDataCollator

data_collator = DefaultDataCollator()
````

## í›ˆë ¨[[train]]

<Tip>

`Trainer`ë¥¼ ì´ìš©í•´ ëª¨ë¸ì„ ë¯¸ì„¸ ì¡°ì •í•˜ëŠ” ê²ƒì— ìµìˆ™í•˜ì§€ ì•Šë‹¤ë©´, [ì—¬ê¸°](https://huggingface.co/docs/transformers/main/ko/tasks/../training#train-with-pytorch-trainer)ì—ì„œ ê¸°ì´ˆ íŠœí† ë¦¬ì–¼ì„ ì‚´í´ë³´ì„¸ìš”!

ì´ì œ ëª¨ë¸ í›ˆë ¨ì„ ì‹œì‘í•  ì¤€ë¹„ê°€ ë˜ì—ˆìŠµë‹ˆë‹¤! `AutoModelForQuestionAnswering`ìœ¼ë¡œ DistilBERTë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤:

````
from transformers import AutoModelForQuestionAnswering, TrainingArguments, Trainer

model = AutoModelForQuestionAnswering.from_pretrained("distilbert-base-uncased")
````

ì´ì œ ì„¸ ë‹¨ê³„ë§Œ ë‚¨ì•˜ìŠµë‹ˆë‹¤:

1. `TrainingArguments`ì—ì„œ í›ˆë ¨ í•˜ì´í¼íŒŒë¼ë¯¸í„°ë¥¼ ì •í•©ë‹ˆë‹¤. ê¼­ í•„ìš”í•œ ë§¤ê°œë³€ìˆ˜ëŠ” ëª¨ë¸ì„ ì €ì¥í•  ìœ„ì¹˜ë¥¼ ì§€ì •í•˜ëŠ” `output_dir` ì…ë‹ˆë‹¤. `push_to_hub=True`ë¡œ ì„¤ì •í•´ì„œ ì´ ëª¨ë¸ì„ Hubë¡œ í‘¸ì‹œí•©ë‹ˆë‹¤ (ëª¨ë¸ì„ ì—…ë¡œë“œí•˜ë ¤ë©´ Hugging Faceì— ë¡œê·¸ì¸í•´ì•¼ í•©ë‹ˆë‹¤).
2. ëª¨ë¸, ë°ì´í„° ì„¸íŠ¸, í† í¬ë‚˜ì´ì €, ë°ì´í„° ì½œë ˆì´í„°ì™€ í•¨ê»˜ `Trainer`ì— í›ˆë ¨ ì¸ìˆ˜ë“¤ì„ ì „ë‹¬í•©ë‹ˆë‹¤.
3. `train()`ì„ í˜¸ì¶œí•´ì„œ ëª¨ë¸ì„ ë¯¸ì„¸ ì¡°ì •í•©ë‹ˆë‹¤.

````
training_args = TrainingArguments(
    output_dir="my_awesome_qa_model",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    push_to_hub=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_squad["train"],
    eval_dataset=tokenized_squad["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
)

trainer.train()
````
