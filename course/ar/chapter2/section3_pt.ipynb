```python
# أول حاجة بنستورد إعدادات وموديل بيرت
from transformers import BertConfig, BertModel

# ننشئ إعدادات جديدة من غير تحميل موديل جاهز
config = BertConfig()

# نبني الموديل باستخدام الإعدادات
model = BertModel(config)

# الموديل هنا لسه متسحبش أوزان مدربة، يعني
# الموديل متهيأ بشكل عشوائي من البداية، مش موديل جاهز أو متدرب
```

# شرح الإعدادات

"attention_probs_dropout_prob": 0.1
ده نسبة الحشو العشوائي اللي بيتطبق على احتمالات الانتباه عشان يمنع الإفراط في التعلّم.

"hidden_act": "gelu"
دالة التفعيل المستخدمة جوه الطبقات المخفية، هنا gelu اللي هي الأفضل عادة لبيرت.

"hidden_dropout_prob": 0.1
نسبة الـ dropout اللي بيتطبق على الطبقات المخفية نفسها.

"hidden_size": 768
عدد الأبعاد في تمثيل كل كلمة داخل النموذج، يعني حجم كل embedding.

"initializer_range": 0.02
النطاق اللي بتبدأ منه أوزان النموذج لما يتم تهيئتها.

"intermediate_size": 3072
حجم الطبقة البين مخفية في الـ Transformer، غالبًا أكبر من hidden_size.

"layer_norm_eps": 1e-12
ثابت صغير للثبات العددي في الطبقات التنسيقية (layer normalization).

"max_position_embeddings": 512
أقصى طول للجملة (عدد الكلمات) اللي يقدر النموذج يتعامل معها.

"model_type": "bert"
نوع النموذج، هنا بيرت.

"num_attention_heads": 12
عدد رؤوس الانتباه المتوازية في كل طبقة.

"num_hidden_layers": 12
عدد طبقات الـ Transformer.

"pad_token_id": 0
رقم رمز الحشو (padding token) المستخدم.

"position_embedding_type": "absolute"
نوع تمثيل المواضع، هنا تمثيل مطلق.

"transformers_version": "4.51.3"
إصدار مكتبة Transformers المستخدم.

"type_vocab_size": 2
عدد أنواع الرموز (token types)، عادة لتمييز بين جملتين (مثلاً في الـ NSP).

"use_cache": true
يعني النموذج يستخدم التخزين المؤقت عشان يسرع التوليد.

"vocab_size": 30522
حجم القاموس اللي بيشتغل عليه النموذج، عدد الكلمات أو الرموز اللي بيعرفها.

 دلوقتي نقدر نحمل الموديل من الجهاز نفسه

```python
from transformers import BertModel
model = BertModel.from_pretrained("./arabic-text-classification")
```

```python
sequences = ["مرحبا!", "رائع.", "جميل!"]
encoded_sequences = [
    [101, 7592, 999, 102],
    [101, 4658, 1012, 102],
    [101, 3835, 999, 102],
]

import torch

model_inputs = torch.tensor(encoded_sequences)

output = model(model_inputs)
```

```python
# نطبع شكل التمثيلات النهائية (last hidden state)
print(output.last_hidden_state.shape)

# نطبع التمثيل الرقمي لأول كلمة في أول جملة كمثال
print(output.last_hidden_state[0][0])
```
